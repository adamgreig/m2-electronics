```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (9, 5)
```

```python
f0 = 434650000.0
t = np.linspace(0, 0.01, 1e8)
lo = np.sin(2*np.pi*f0*t)
```

```python
plt.plot(t[-100:], lo[-100:])
plt.title("LO @ 434.650M")
```

```python
freqs = np.repeat(f0, 1e8)
carrier = np.sin(2*np.pi*np.cumsum(freqs) / (1e8 / 0.01))
plt.plot(t[-100:], carrier[-100:])
plt.title("Unmodulated Carrier @ 434.650M")
```

```python
freqs[:0.9e8-10] = f0 * 1.8
mod_carrier = np.sin(2*np.pi*np.cumsum(freqs) / (1e8 / 0.01))
plt.plot(t[-100:], mod_carrier[-100:])
plt.title("Shifted Carrier @ 434.650M")
```
