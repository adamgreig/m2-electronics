```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

We expect the resistance $R$ of the Metron to be within $0.9\Omega$ to $1.6\Omega$ according to the datasheet, but let's validate that we get a desired current over a wider range.

```python
R = np.linspace(0.6, 2.0, 100)
```

The battery voltage is nominally $1.5V$ and is approximately flat until complete discharge. We have four such in a pack, so assume a nominal $V_b$ of $6V$.

```python
Vb = 6.0
```

The bias resistor $R_B$ is $5\Omega\pm5\%$, i.e. $4.95\Omega$ to $5.05\Omega$.

```python
RB = np.linspace(4.95, 5.05, 10)
```

Current through the pyro is $I=V/R=\frac{V_b}{R+R_B}$

```python
plt.figure(figsize=(12, 6))
for rb in RB:
    I = Vb / (R + rb)
    plt.plot(R, I, color='b')
plt.xlim((0.6, 2.0))
plt.ylim((0.5, 1.2))
plt.axhline(1.0, color='g', label="Recommended Firing Current")
plt.axhline(0.6, color='r', label="Minimum Firing Current")
plt.legend()
```

Should be good.
